```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Iris dataset directly from seaborn
df = sns.load_dataset('iris')

# Basic structure
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nFirst 5 rows:")
df.head()
```

```python
print("Missing Values:\n", df.isnull().sum())
print("\nSpecies Count:\n", df['species'].value_counts())
print("\nStatistical Summary:")
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', 
                hue='species', palette='Set1', s=80)
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.tight_layout()
plt.savefig('scatter_plot.png', dpi=150)
plt.show()
```

```python
df.hist(figsize=(10,6), bins=20, color='steelblue', edgecolor='black')
plt.suptitle('Feature Distribution - Iris Dataset', y=1.02)
plt.tight_layout()
plt.savefig('histogram.png', dpi=150)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,7))

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    ax = axes[i//2, i%2]
    sns.boxplot(data=df, x='species', y=feature, 
                palette='Set2', ax=ax)
    ax.set_title(f'{feature} by Species')
    ax.set_xlabel('')

plt.suptitle('Feature Distribution by Species - Box Plots', y=1.02)
plt.tight_layout()
plt.savefig('boxplot.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(df.drop(columns='species').corr(), 
            annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.savefig('heatmap.png', dpi=150)
plt.show()
```
